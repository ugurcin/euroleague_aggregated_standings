```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
urls = ["https://en.wikipedia.org/wiki/2022%E2%80%9323_EuroLeague_Regular_Season",
        "https://en.wikipedia.org/wiki/2021%E2%80%9322_EuroLeague_Regular_Season",
        "https://en.wikipedia.org/wiki/2020%E2%80%9321_EuroLeague_Regular_Season", 
        "https://en.wikipedia.org/wiki/2019%E2%80%9320_EuroLeague_Regular_Season",
        "https://en.wikipedia.org/wiki/2018%E2%80%9319_EuroLeague_Regular_Season",
        "https://en.wikipedia.org/wiki/2017%E2%80%9318_EuroLeague_Regular_Season",
        "https://en.wikipedia.org/wiki/2016%E2%80%9317_EuroLeague_Regular_Season"]
table_class = "wikitable sortable jquery-tablesorter"
dfs = []
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    euro_table = soup.find('table', {'class':"wikitable"})
    df_new = pd.read_html(str(euro_table))
    df_new = pd.DataFrame(df_new[0])
    df_new = df_new.iloc[ : , :7]
    df_new.columns = ['Pos', 'Team', 'Pld', 'W', 'L', 'PF', 'PA']
    dfs.append(df_new)
```

```python
df = pd.concat(dfs)
df = df[df['Pld'] !=0]
```
